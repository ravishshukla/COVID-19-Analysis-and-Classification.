# Detecting COVID-19 with Chest X Ray using PyTorch

Image classification of Chest X Rays in one of three classes: Normal, Viral Pneumonia, COVID-19

Dataset from [COVID-19 Radiography Dataset](https://www.kaggle.com/tawsifurrahman/covid19-radiography-database) on Kaggle

# Importing Libraries

```python
%matplotlib inline

import os
import shutil
import random
import torch
import torchvision
import numpy as np

from PIL import Image
from matplotlib import pyplot as plt

torch.manual_seed(0)

print('Using PyTorch version', torch.__version__)
print("Imported!!!")
```

# Preparing Training and Test Sets

```python
class_names = ['normal', 'viral', 'covid']
root_dir = 'COVID-19 Radiography Database'
source_dirs = ['NORMAL', 'Viral Pneumonia', 'COVID-19']

if os.path.isdir(os.path.join(root_dir, source_dirs[1])):
    os.mkdir(os.path.join(root_dir, 'test'))

    for i, d in enumerate(source_dirs):
        os.rename(os.path.join(root_dir, d), os.path.join(root_dir, class_names[i]))

    for c in class_names:
        os.mkdir(os.path.join(root_dir, 'test', c))

    for c in class_names:
        images = [x for x in os.listdir(os.path.join(root_dir, c)) if x.lower().endswith('png')]
        selected_images = random.sample(images, 30)
        for image in selected_images:
            source_path = os.path.join(root_dir, c, image)
            target_path = os.path.join(root_dir, 'test', c, image)
            shutil.move(source_path, target_path)
print("Improted!!!")
```

# Creating Custom Dataset

```python
class ChestXRayDataset(torch.utils.data.Dataset): #Subclass from pytorch
    def __init__(self, image_dirs, transform): #Initializer fuction
        def get_images(class_name):
            images = [x for x in os.listdir(image_dirs[class_name]) if x.lower().endswith('png')]
            print(f'Found {len(images)} {class_name} examples')
            return images
        
        self.images = {}
        self.class_names = ['normal','viral','covid']
        
        for c in self.class_names:
            self.images[c] = get_images(c)
        
        self.image_dirs = image_dirs
        self.transform = transform
            
            
    def __len__(self):
        return sum([len(self.images[c]) for c in self.class_names])
    
    
    def __getitem__(self, index):
        class_name = random.choice(self.class_names)
        index = index % len(self.images[class_name]) # this is done because covid dataset has lesser no. of examples from normal and viral, so if an index comes to be 1300 it will be stisfied by the normal or viral by not by covid, so we take the remainder of the index when divided index when divided by the length of the images in that class
        image_name = self.images[class_name][index]
        image_path = os.path.join(self.image_dirs[class_name], image_name)
        image = Image.open(image_path).convert('RGB') #Image belongs to Pillow module(PIL) #we have to specify RGB because we are going to use Resnet18's pretrained weights wihch are trained from the "ImageNet" dataset
        return self.transform(image),self.class_names.index(class_name)
```

# Image Transformations

```python
train_transform = torchvision.transforms.Compose([
    torchvision.transforms.Resize(size=(224,224)), #covert to a size that our model recognizes
    torchvision.transforms.RandomHorizontalFlip(), #Data augmentation
    torchvision.transforms.ToTensor(), # Because pytorch dos not recognize pIL images, it recognizes Tensors
    torchvision.transforms.Normalize(mean = [0.485, 0.456, 0.406], #we normalize ot to specific range for the 3 Channels(RGB) which was used when Resnet18 model was trained on ImageNet Data set to benifit from the Transfer Learning
                                    std = [0.229,0.224,0.225])

])
```

```python
test_transform = torchvision.transforms.Compose([
    torchvision.transforms.Resize(size=(224,224)), #covert to a size that our model recogniz
    torchvision.transforms.ToTensor(), # Because pytorch dos not recognize pIL images, it recognizes Tensors
    torchvision.transforms.Normalize(mean = [0.485, 0.456, 0.406], #we normalize ot to specific range for the 3 Channels(RGB) which was used when Resnet18 model was trained on ImageNet Data set to benifit from the Transfer Learning
                                    std = [0.229,0.224,0.225])

])
```

# Prepare DataLoader

```python
train_dirs = {
    'normal': 'COVID-19 Radiography Database/normal',
    'viral': 'COVID-19 Radiography Database/viral',
    'covid': 'COVID-19 Radiography Database/covid'
}

train_dataset = ChestXRayDataset(train_dirs, train_transform)
```

```python
test_dirs = {
    'normal': 'COVID-19 Radiography Database/test/normal',
    'viral': 'COVID-19 Radiography Database/test/viral',
    'covid': 'COVID-19 Radiography Database/test/covid'
}

test_dataset = ChestXRayDataset(test_dirs, test_transform)
```

```python
batch_size = 6

dl_train = torch.utils.data.DataLoader(train_dataset, batch_size = batch_size,
                                      shuffle=True)
dl_test = torch.utils.data.DataLoader(test_dataset, batch_size= batch_size, shuffle=True)


print("Number of traing batches ", len(dl_train))
print("Number of Test Batches ", len(dl_test))
```

# Data Visualization

```python
class_names = train_dataset.class_names


def show_images(images,lables,preds):
    plt.figure(figsize=(8,4))
    
    for i, image in enumerate(images):
        plt.subplot(1,6, i + 1, xticks=[],yticks=[])
        image = image.numpy().transpose((1, 2, 0)) # transpose in such a way that the first axis is the last becaus  the pytorch implimentation is channel first then height then the width
        mean = np.array([0.485, 0.456, 0.406])
        std = np.array([0.229,0.224,0.225])
        image = image *std +mean #convert imgae pixel value to the state before normalization
        image = np.clip(image, 0., 1.)
        plt.imshow(image)
        
        col = 'green' if preds[i] == labels[i] else 'red'
        
        plt.xlabel(f'{class_names[int(labels[i].numpy())]}')
        plt.ylabel(f'{class_names[int(preds[i].numpy())]}', color = col)
        
    plt.tight_layout()
    plt.show()
```

```python
images,labels = next(iter(dl_train))
show_images(images, labels, labels)
```

```python
images, labels = next(iter(dl_test))
show_images(images, labels, labels)
```

# Creating the Model

```python
resnet18 = torchvision.models.resnet18(pretrained=True)
print(resnet18)
```

```python
resnet18.fc = torch.nn.Linear(in_features=512,out_features=3)
loss_fn = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(resnet18.parameters(), lr = 3e-5) # Adam works of stochastic grad descent, and lr is the learning rate
```

```python
def show_preds():
    resnet18.eval()
    images, labels = next(iter(dl_test))
    outputs = resnet18(images)
    _, preds = torch.max(outputs, 1)#torch.max( returns values and the index for all the outputs, we dont want the value just the index so we use, (_, preds) syntax
    show_images(images,labels,preds)
```

```python
show_preds()
```

# Training the Model

```python
def train(epochs):
    print('Strating traing...')
    for e in range(0, epochs):
        print('='*20)
        print(f'Starting epoch {e+1}/{epochs}')
        print('='*20)
        
        train_loss =0
        
        resnet18.train()
        
        for train_step, (images, labels) in enumerate(dl_train):
            optimizer.zero_grad()
            outputs = resnet18(images)
            loss = loss_fn(outputs, labels)
            loss.backward() #this is the back propagation step
            optimizer.step()
            train_loss += loss.item()
            if train_step %20 ==0:
                print('Eavaluating Modelat step', train_step)
                acc = 0
                val_loss = 0
                resnet18.eval()
                
                for val_step, (images, labels) in enumerate(dl_test):
                    outputs = resnet18(images)
                    loss = loss_fn(outputs, labels)
                    val_loss += loss.item()
                            
                    _, preds = torch.max(outputs, 1) #Dimension 1 corresponds to the values and 2 corresponds to the examples
                    acc += sum((preds == labels).numpy())
                val_loss /= (val_step +1)
                acc = acc/ len(test_dataset)
                print(f'Val loss; {val_loss:.4f}, Acc: {acc:.4f}')
                show_preds()
                
                resnet18.train()
                
                if acc> 0.95:
                    print('Performance condition Satisfied...')
                    return
                
        train_loss /= (train_step + 1)
        print(f'Training loss : {train_loss: .4f}')
```

```python
train(epochs=2)
```
